# Data preparation

```python
train_copy['Dependents'].replace('3+',3 ,inplace=True)
test['Dependents'].replace('3+',3 ,inplace=True)
train_copy['Loan_Status'].replace('Y',1 ,inplace=True)
train_copy['Loan_Status'].replace('N',0 ,inplace=True)
train_copy['Gender'].replace('Male',1 ,inplace=True)
train_copy['Gender'].replace('Female',0 ,inplace=True)
train_copy['Married'].replace('Yes',1 ,inplace=True)
train_copy['Married'].replace('No',0 ,inplace=True)
train_copy['Education'].replace('Graduate',1 ,inplace=True)
train_copy['Education'].replace('Not Graduate',0 ,inplace=True)
train_copy['Self_Employed'].replace('No',0 ,inplace=True)
train_copy['Self_Employed'].replace('Yes',1 ,inplace=True)
train_copy['Property_Area'].replace('Rural',1 ,inplace=True)
train_copy['Property_Area'].replace('Urban',0 ,inplace=True)
train_copy['Property_Area'].replace('Semiurban',2 ,inplace=True)


train_copy

```

```python
train_copy["TotalIncome"]=train_copy["ApplicantIncome"]+train_copy["CoapplicantIncome"]
test_copy["TotalIncome"]=test_copy["ApplicantIncome"]+test_copy["CoapplicantIncome"]


train_copy["Loan_per_mounth"]=train_copy["LoanAmount"]/train_copy["Loan_Amount_Term"]
test_copy["Loan_per_mounth"]=test_copy["LoanAmount"]/test_copy["Loan_Amount_Term"]

train_copy["Income_after_debt"] = train_copy["TotalIncome"]-train_copy["Loan_per_mounth"]
test_copy["Income_after_debt"] = test_copy["TotalIncome"]-test_copy["Loan_per_mounth"]

train_copy
```

```python
train_copy=train_copy.drop(["ApplicantIncome","CoapplicantIncome","LoanAmount","Loan_Amount_Term",'Loan_ID'],axis=1)
test_copy = test_copy.drop(["ApplicantIncome","CoapplicantIncome","LoanAmount","Loan_Amount_Term",'Loan_ID'],axis=1)
```

# Model Building

```python
X=train_copy.drop('Loan_Status',1)
Y=train_copy['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_cv,y_train,y_cv=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
logistic_model = LogisticRegression(random_state=1)
logistic_model.fit(x_train,y_train)
pred_cv_logistic=logistic_model.predict(x_cv)
score_logistic =accuracy_score(pred_cv_logistic,y_cv)*100 
score_logistic
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc


# Calculate ROC curve metrics
fpr, tpr, thresholds = roc_curve(y_cv, pred_cv_logistic)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure()
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--', label='No Skill')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve for Logistic Regression Model')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix

# Create the confusion matrix
cm = confusion_matrix(y_cv, pred_cv_logistic)

# Print the confusion matrix
print(cm)

# Optionally, create a visual representation using matplotlib
import matplotlib.pyplot as plt

plt.matshow(cm)
plt.colorbar()
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
